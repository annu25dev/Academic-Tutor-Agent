```python
def calculate_average_latency(latencies):
    """
    Calculate average response latency.

    Args:
        latencies (list): List of response times in seconds.

    Returns:
        float: Average latency in seconds.
    """
    return sum(latencies) / len(latencies)


# Response latency values from test cases
latencies = [4.39, 3.24, 2.71, 2.82, 7.30, 6.45, 5.27, 5.36, 13.14, 3.90]

average_latency = calculate_average_latency(latencies)

print(f"Average Response Latency: {average_latency:.2f}s")
```

## DeepEval Faithfulness Score (For RAG)

```python
#IF JSONDecodeError comes, please run the code again 
import os
#os.chdir("Backend") #-Run this once before testing
print(os.getcwd())
import nest_asyncio
nest_asyncio.apply()
from crewai import Crew, Process
import asyncio
from concurrent.futures import ThreadPoolExecutor

from deepeval.test_case import LLMTestCase
from deepeval.metrics import FaithfulnessMetric
from deepeval.models import DeepEvalBaseLLM

from dotenv import load_dotenv

from groq import Groq
 
from patches.groq_patch import apply_patch
apply_patch()

from RAG_agent.Retrieval_pipeline import retrieve_context
from crew import run_academic_tutor
```

```python
load_dotenv()
GROQ_KEY = os.getenv("GROQ_API_KEY")
client = Groq(api_key=GROQ_KEY)


class GroqEvaluator(DeepEvalBaseLLM):

    def __init__(self):
        self.model_name = "llama-3.3-70b-versatile"
        self.client = client


    def load_model(self):
        return self.client


    def generate(self, prompt: str):

        response = self.client.chat.completions.create(
            model=self.model_name,
            messages=[
                {
                    "role": "user",
                    "content": prompt
                }
            ],
            temperature=0
        )

        return response.choices[0].message.content


    async def a_generate(self, prompt: str):
        return self.generate(prompt)


    def get_model_name(self):
        return self.model_name



evaluator = GroqEvaluator()
```

```python
student_queries = [
    "What are the key leadership qualities discussed in the notes?",
    "Explain the leadership styles mentioned in the document.",
    "What are the three barriers to critical thinking?",
    "How can critical thinking barriers be overcome?",
    "Explain Bloom's Taxonomy levels.",
    "What is the difference between Analysis and Evaluation in Bloom's Taxonomy?",
    "What are the basic rules of netiquette?",
    "What are the important aspects of professional grooming?",
    "What topics are included in the uploaded syllabus?",
    "Summarize the uploaded syllabus."
]
#RUNNING THE TEST CASES ONE BY ONE TO AVOID TOKEN LIMIT ERROR
student_query=student_queries[1]  #Change here 

student_id = "deepeval_test_student"
session_id = "deepeval_test_session"
```

```python
rag_output = retrieve_context(student_query)

retrieved_context = rag_output["context"]


print("Retrieved Context:")
print(retrieved_context)
```

```python

executor = ThreadPoolExecutor(max_workers=1)

loop = asyncio.get_running_loop()

tutor_answer = await loop.run_in_executor(
    executor,
    lambda: run_academic_tutor(
        student_id=student_id,
        session_id=session_id,
        student_question=student_query
    )
)

print(tutor_answer)
```

```python
test_case = LLMTestCase(
    input=student_query,
    actual_output=tutor_answer,
    retrieval_context=[
        retrieved_context
    ]
)
```

```python
faithfulness = FaithfulnessMetric(
    threshold=0.85,
    model=evaluator
)


score = faithfulness.measure(test_case)


print("Faithfulness Score:", faithfulness.score)

print("\nReason:")
print(faithfulness.reason)
```
